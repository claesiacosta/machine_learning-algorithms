# K Nearest Neighbors - Projeto 

```python
#Importando pandas, seaborn e as bibliotecas usuais
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Obtendo os dados


```python
df = pd.read_csv("KNN_Project_Data")
```

```python
# Verificando cabeçalho
df.head()
```

Usando seaborn no DataFrame para criar um pairplot para coluna TARGET CLASS.

# Padronize as variáveis

Padronizar as variáveis.


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler  = StandardScaler() # instânciando
```

```python
scaler.fit(df.drop('TARGET CLASS', axis = 1)) # método fit() para treinar o modelo
```

```python
param_padronizados = scaler.transform(df.drop('TARGET CLASS', axis = 1)) #tranformando param em uma versao padronizada
```

```python
df_param = pd.DataFrame(param_padronizados, columns = df.columns[:-1]) # convertendo em dataframe
df.head()
```

# Divisão treino-teste


```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_param, df['TARGET CLASS'], test_size=0.3) 
#dividindo em conjunto de treino e de teste
```

# Usando o KNN


```python
from sklearn.neighbors import KNeighborsClassifier #importando classificador
```

```python
knn = KNeighborsClassifier(n_neighbors=1) # criando instância do modelo com os vizinhos
```

```python
knn.fit(X_train, y_train) # ajustando para os dados de treino
```

# Previsões e avaliações


```python
pred = knn.predict(X_test) #usando método de previsão
```

```python
from sklearn.metrics import classification_report, confusion_matrix # importando relatório de classificação
# e matrix de confusão
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

# Escolhendo o valor K

```python
# loop for para treinar vários modelos KNN com valores k diferentes e, 
#em seguida, mantendo um registro do error_rate para cada um desses modelos com uma lista.
error_rate = []

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    error_rate.append(np.mean(pred!=y_test))
error_rate
```

```python
# gráfico para visualização dos dados
plt.figure(figsize=(14,8))
plt.plot(range(1,40), error_rate, color='blue', linestyle='dashed', marker='o', markersize=10, markerfacecolor='red')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.title('Error Rate vs. K Value')
```

## Treinando modelo novamente com novo valor K


```python
# treinado modelo com o novo valor K e refazendo relatório e matrix para treino
knn = KNeighborsClassifier(n_neighbors=30)

knn.fit(X_train, y_train)
pred = knn.predict(X_test)

print('WITH K=30')
print('\n')
print(confusion_matrix(y_test, pred))
print('\n')
print(classification_report(y_test, pred))
print('\n')
```
