```python
df = pd.read_csv("Classified Data", index_col=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('TARGET CLASS',axis=1))
```

```python
df_normalizado = scaler.transform(df.drop('TARGET CLASS',axis=1))
```

```python
df_normalizado
```

```python
df_param = pd.DataFrame(df_normalizado, columns = df.columns[:-1])
```

```python
df_param
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_param, df['TARGET CLASS'], test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1) # parametro é o número de vizinhos
```

```python
knn
```

```python
knn.fit(X_train, y_train)
```

```python
predict = knn.predict(X_test)
```

```python
predict
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predict))
```

```python
print(confusion_matrix(y_test, predict))
```

```python
error_rate = []

for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    predict = knn.predict(X_test)
    error_rate.append(np.mean(predict != y_test))
```

```python
plt.figure(figsize=(14, 8))
plt.plot(range(1, 40),error_rate, color='red', linestyle='dashed', marker='o')
```

```python
knn = KNeighborsClassifier(n_neighbors = 30)
knn.fit(X_train, y_train)
predict = knn.predict(X_test)
```

```python
print(classification_report(y_test, predict))
```
