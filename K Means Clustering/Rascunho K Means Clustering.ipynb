```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=50)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(data[0][:,0],data[0][:,1], c = data[1], cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4) # n_clusters Ã© a quantidade de grupos
```

```python
kmeans.fit(data[0]) # treinamento
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey = True, figsize=(12, 8))
ax1.set_title("KMeans")
ax1.scatter(data[0][:,0],data[0][:,1], c = kmeans.labels_, cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1], c = data[1], cmap='rainbow')
```
